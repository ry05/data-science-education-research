```python
import os
import pandas as pd

from gds_operations import CourseScorer, ProgramScorer, ScoreNormalizer
```

```python
label_dictionary = pd.read_csv('labelled_units_words.csv')

data_path = "../../data_collection/data/processed"
pgm_india = pd.read_csv(os.path.join(data_path, "india_pgms.csv")).fillna("Not inferred")
pgm_usa = pd.read_csv(os.path.join(data_path, "usa_pgms.csv")).fillna("Not inferred")
curr_india = pd.read_csv(os.path.join(data_path, "india_curr.csv")).fillna("Not inferred")
curr_usa = pd.read_csv(os.path.join(data_path, "usa_curr.csv")).fillna("Not inferred")
curr_usa = curr_usa.drop(['Unnamed: 4', 'Unnamed: 5'], axis=1)

india_data = curr_india.merge(pgm_india)
usa_data = curr_usa.merge(pgm_usa)
```

```python
df = pd.concat([curr_usa, curr_india])
df = df.drop(['course outcome or overview', 'topics covered'], axis=1)
df.shape
```

```python
df.head()
```

```python
cs = CourseScorer(label_dictionary, df)
t = cs.score_courses_by_gds()
t.head(10)
```

```python
def label_by_subject(x, subject):

    if subject in x.lower():
        return 1
    return 0

t['Data Gathering, Preparation and Exploration'] = t['Data Gathering, Preparation and Exploration'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "data science"))
t['Data Representation and Transformation'] = t['Data Representation and Transformation'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "data science"))
t['Computing with Data'] = t['Computing with Data'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "data science"))
t['Data Modeling'] = t['Data Modeling'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "data science"))
t['Data Visualization and Presentation'] = t['Data Visualization and Presentation'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "data science"))


t['Data Gathering, Preparation and Exploration'] = t['Data Gathering, Preparation and Exploration'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "artificial intelligence"))
t['Data Representation and Transformation'] = t['Data Representation and Transformation'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "artificial intelligence"))
t['Computing with Data'] = t['Computing with Data'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "artificial intelligence"))
t['Data Modeling'] = t['Data Modeling'] | t['compulsory course'].apply(lambda x: label_by_subject(x, "artificial intelligence"))
```

```python
sn = ScoreNormalizer(t)
t2 = sn.update_scored_df()
t2.head()
```
