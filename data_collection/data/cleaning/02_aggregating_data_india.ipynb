# Aggregating Data - India

```python
import sys
sys.path.insert(0, "scraping")

import pandas as pd
```

```python
ppl = pd.read_excel("data/PPL/PPL_India.xlsx")
pgm_descr = pd.read_csv("data/intermediate/program_description_india.csv")
pgm_other = pd.read_csv("data/intermediate/program_other_data_india.csv")
```

```python
ppl.columns
```

```python
ppl = ppl.drop(['random comments', 'Type of website', 'Useful area', 'How is curriculum data displayed?'], axis=1)
pgm_descr = pgm_descr.drop(['header'], axis=1)
```

```python
temp1 = ppl.merge(pgm_descr, left_on='URL', right_on='url', validate='1:1', how='outer')
temp1 = temp1.drop('URL', axis=1)

temp1.head(2)
```

```python
temp2 = temp1.merge(pgm_other, on='url', how='outer')
temp2.head()
```

```python
ren_cols = [
    'uni_name',
    'pgm_name',
    'dur_yrs',
    'type',
    'national_rank_qs',
    'dept_involved',
    'location',
    'fees_inr',
    'url',
    'descr',
    'header_tag',
    'header_names',
    'links',
    'date_pub',
    'date_last_mod',
    'html_tags',
    'html_tag_freq',
    'html_emphasized_elements'
]

temp2.columns = ren_cols
```

```python
temp2.info()
```

```python
temp2.head()
```

```python
# remove fee information
temp2 = temp2.drop('fees_inr', axis=1)
```
