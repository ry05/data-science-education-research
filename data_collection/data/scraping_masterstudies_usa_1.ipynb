```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait       
from selenium.webdriver.common.by import By       
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
# main URL
masterstudies = "https://www.masterstudies.com/universities/USA/"
```

```python
programs = []
```

```python
# open the URL using the selenium driver
driver = webdriver.Chrome("chromedriver_win32/chromedriver.exe")
driver.get(masterstudies)
time.sleep(5)
print("Opened home webpage")

# need to click on "programs"
programs_tab_xpath = '//*[@id="main"]/div/div[5]/div/div/div[2]'
programs_tab = driver.find_element_by_xpath(programs_tab_xpath)
driver.execute_script("arguments[0].click();", programs_tab)
programs_tab.click()
time.sleep(5)
print("Moved into Programs tab")

# program list
#programs_xpath = '//*[@id="main"]/div/div[6]/div/div[2]/div[2]/div[2]'
#programs = driver.find_element_by_class_name('program-listitem')
#programs = driver.find_elements_by_class_name('program-listitem')

soup = BeautifulSoup(driver.page_source, 'html.parser')
progs = soup.find_all('div', class_ = 'program-listitem')

program = [prog for prog in progs]
programs += program

next_page_btn_xpath = '//*[@id="main"]/div/div[6]/div/div[3]/a/button'
btn = driver.find_element_by_xpath(next_page_btn_xpath)
driver.execute_script("arguments[0].click();", btn)
print("Moving to next page")
```

```python
num_pages = 297
for page in range(1, num_pages):
    print(page)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    progs = soup.find_all('div', class_ = 'program-listitem')

    program = [prog for prog in progs]
    programs += program

    if page != (num_pages-1):
        # next page
        print(f'Printing page {page+1}')
        next_page_btn_xpath = '//*[@id="main"]/div/div[6]/div/div[3]/a[2]/button'
        btn = driver.find_element_by_xpath(next_page_btn_xpath)
        driver.execute_script("arguments[0].click();", btn)
        #driver.implicitly_wait(10)
        time.sleep(5)
        print("Moving to next page")

#print(len(titles))
```

```python
def make_data(program_tags):
    """
    Make the data from the program tags
    """ 

    titles = []
    url = []
    school = []
    degree = []
    pace = []
    duration = []
    languages = []
    based = []
    price = []

    for pgm in range(len(program_tags)):

        title = program_tags[pgm].select('.title')[0].text.strip().replace('\n', '').replace('\xa0', '')
        '''
        if any (word in title.lower() for word in KEYWORDS):
            titles.append(title)
        '''
        titles.append(title)

        try:
            url.append(programs[pgm].select('.title')[0].a['href'].strip().replace('\n', ''))
        except:
            url.append(np.nan)

        try:
            school.append(program_tags[pgm].select('.school')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            school.append(np.nan)

        try:
            degree.append(program_tags[pgm].select('.degree')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            degree.append(np.nan)

        try:
            pace.append(program_tags[pgm].select('.pace')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            pace.append(np.nan)

        try:
            duration.append(program_tags[pgm].select('.duration')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            duration.append(np.nan)
            
        try:
            languages.append(program_tags[pgm].select('.languages')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            languages.append(np.nan)

        try:
            based.append(program_tags[pgm].select('.based')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            based.append(np.nan)

        try:
            price.append(program_tags[pgm].select('.price')[0].text.strip().replace('\n', '').replace('\xa0', ''))
        except:
            price.append(np.nan)

    df = pd.DataFrame(
        dict(
            title = titles,
            url = url,
            school = school,
            degree = degree,
            pace = pace,
            duration = duration,
            languages = languages,
            based = based,
            price = price
        )
    )

    print(df.shape)

    return df
        
```

```python
temp_df = make_data(programs)
```
